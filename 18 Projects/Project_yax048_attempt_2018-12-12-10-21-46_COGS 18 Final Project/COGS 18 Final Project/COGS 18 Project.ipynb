# Reservation-Making ChatBot
For this project, I will write a code for a chatbot which could help people to:
    1. make appointments
    2. check appointments
    3. cancel appointments.

I create a class called reservation list which include methods to add, check and cancel appointments.

I started with a dictionary called availiable which includes all avaliable times in it, and an empty dictionary called appointments which will includes all appointment times user make.

    The add appointment method could move one key(with the time) from availiable dictionary to appointment dictionary, and adding job name.
    The check appointment method could print out the item of a specific key which user enter.
    The cancel appointment method could move one key(with the time) from appointment dictionary to availiable dictionary, and delete the job name. 

Then I create a chatbot use methods above to approch my goal.

```python
# import class reservation list and give abbreviate variable name p
from my_module.reservation import reservation_list as p
# sign new variable name r for the class
r = p()
# import funtion while_loop_answer
from my_module.while_loop import while_loop_answer
# import funtions question1 and add_appoint
from my_module.add_appointment import question1,add_appoint
```

```python
def have_a_chat_test():
    """Main function to run our chatbot."""
    chat = True
    while chat:
        # Ask user the question
        question1()
        # Get a answer from the user
        msg1 = input('INPUT :\t')
        
        # If the answer is 1 help user to make appointment 
        if msg1 == '1':
            # If there is no availiable time, report no availiable time to user
            if r.availiable == {}:
                print("Sorry, we don't have any time avaliable now")
            else:
                add_appoint()
            # After making the appointment, ask user if they need anything else   
            print('Anything else I can help you with? (please answer Yes or No)')
            # If the answer is unvalid, keep report error until the answer is valid
            chat = while_loop_answer()

        # If the answer is 2 help user to check the appointment 
        elif msg1 == '2':
            print('May I have your appointment# please?')
            msg1_2 = input('INPUT :\t')
            r.check_appointment(msg1_2)
            print('Anything else I can help you with? (please answer Yes or No)')
            # If the answer is unvalid, keep report error until the answer is valid
            chat = while_loop_answer()

        # If the answer is 3 help user to cancel an appointment 
        elif msg1 == '3':
            print('May I have your appointment# please?')
            msg1_3 = input('INPUT :\t')
            r.cancel_appointment(msg1_3)
            print('Anything else I can help you with? (please answer Yes or No)')
            # If the answer is unvalid, keep report error until the answer is valid
            chat = while_loop_answer()

        else: 
            # If the answer is unvalid, keep report error until the answer is valid
            print("Sorry, I can't understand you. please call (888)888-8888")
            print('Anything else I can help you with? (please answer Yes or No)')
            chat = while_loop_answer()
```
